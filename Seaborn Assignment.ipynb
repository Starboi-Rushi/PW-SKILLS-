# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Plots of Seaborn:-

1. Histograms:

Histograms are used to visualize the distribution of a single numeric variable.
Histograms used to understand the central tendency, spread, shape of data, making it helpful to understand identify patterns, outliers, or potential data issues.

2. Scatter Plots:

Scatter plots are used to visualize the relationship between two numeric variables.
Scatter plots help you understand the correlation or association between two variables. They are useful for identifying trends, clusters, outliers, and understanding the strength and direction of the relationship.

3. Bar Plots:

Bar plots are suitable for comparing categorical data, such as counts, proportions, or averages, across different categories. They are commonly used for visualizing survey results, market shares, and more.

4. Box Plots (Box-and-Whisker Plots):
 
Box plots are useful for identifying the spread of data, the presence of outliers, and comparing distributions across different categories or groups. They are often used to visualize its summary statistics, including median, quartiles, and potential outliers.
5. Violin Plots:

Violin plots are helpful when you want to compare the distributions of multiple categories and understand their shapes, central tendencies, and variations. They are particularly useful when you have overlapping or multimodal distributions.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
a=sns.load_dataset("fmri")
a
```

```python
plt.figure(figsize=(20, 12))
sns.lineplot(data=a, x = "timepoint" , y = "signal",style="event", hue="region")
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
b=sns.load_dataset("titanic")
b
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=b, x= "pclass", y="age")
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=b, x= "pclass", y="fare")
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
c=sns.load_dataset("diamonds")
c
```

```python
plt.figure(figsize=(15,10))

sns.histplot(data=c, x='price',hue='cut',kde=True)

plt.title('Price of diamond with different cut')
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
d = sns.load_dataset('iris')
d
```

```python
sns.pairplot(data= d,hue="species")
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
e = sns.load_dataset('flights')
e
```

```python
e_pivot = e.pivot_table(index='month', columns='year', values='passengers')
e_pivot
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(e_pivot, cmap="viridis", annot=True, fmt="d",linewidths=0.9)
```
