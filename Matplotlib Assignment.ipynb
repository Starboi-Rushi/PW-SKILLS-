# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a python library used to create 2D graphs and plots by using python scripts. It has a module named pyplot which makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc. It supports a very wide variety of graphs and plots namely - histogram, bar charts, power spectra, error charts etc. It is used along with NumPy to provide an environment that is an effective open source alternative for MatLab. 

### Plots used in matplotlib:-

1. Line Plot: Line plots are used to visualize data points connected by straight lines. They are useful for showing trends over time or relationships between two variables.

2. Bar Plot: Bar plots are used to display categorical data with rectangular bars, where the length of each bar represents a value associated with a category. They are commonly used for comparing categories or showing counts.

3. Scatter Plot: Scatter plots are used to display individual data points as markers on a two-dimensional plane. They are suitable for visualizing the relationship between two variables or identifying outliers.

4. Histogram: Histograms are used to represent the distribution of a continuous variable by dividing it into bins and showing the frequency or count of data points within each bin.

5. Pie Chart: Pie charts are used to display parts of a whole, where each slice represents a portion or percentage of the total. They are effective for visualizing the composition of data.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

###  plot a scatter plot.
### import numpy as np
### np.random.seed(3)
### x = 3 + np.random.normal(0, 2, 50)
### y = 3 + np.random.normal(0, 2, len(x))
### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sizes =3000* np.random.rand(50)
colours = np.random.rand(50)
plt.scatter(x,y, c=colours,s= sizes )
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter Plot X vs Y')
plt.show()
```

The subplot() function in Matplotlib is used to create multiple subplots within the same figure. It allows us to arrange and display multiple plots or charts in a grid-like fashion, which is especially useful when we want to compare different datasets or visualize multiple aspects of our data in the same figure.

```python
import numpy as np
```

```python

plt.figure(figsize=(10,8))

# for Line 1
plt.subplot(2, 2,1)

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

plt.plot(x1, y1)
plt.title('Line 1')

# for Line 2
plt.subplot(2, 2, 2)

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Plot Line 2
plt.plot(x2, y2)
plt.title('Line 2')

#for Line 3
plt.subplot(2, 2, 3)

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Plot Line 3
plt.plot(x3, y3)
plt.title('Line 3')

# for Line 4
plt.subplot(2, 2, 4)

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plot Line 4
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

### import numpy as np
### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
### profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a data visualization technique that uses rectangular bars to represent and compare data values. Each bar typically corresponds to a category or group, and the height or length of the bar represents the value of a variable associated with that category. 

Bar plots are used to display and compare discrete categories and their associated values.

### Uses of bar plots:

1. Comparison: Bar plots are used to compare the values of different categories or groups. They make it easy to visually identify which category has the highest or lowest value.

2. Categorical Data: Bar plots are particularly useful for visualizing categorical data, such as product sales, company profits, survey responses by category, and more.

3. Part-to-Whole Relationships: They can be used to show the composition of a whole by breaking it down into its constituent parts, such as market share percentages.

4. Trends and Comparisons: Bar plots can also be used to visualize trends over time when the x-axis represents time periods.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(10, 8)) 

plt.bar(company, profit, color='#1c9925')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')


plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 8)) 

plt.barh(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.grid()
plt.tight_layout()

plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

### box1 = np.random.normal(100, 10, 200)
### box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistics and characteristics of the data, including the median, quartiles (25th and 75th percentiles), and potential outliers. 

### Box plots are used for the following purposes:

1. Identifying Outliers: Box plots help to identify outliers in the data. Any data points that fall outside the "whiskers" (lines extending from the box) are considered outliers.

2. Comparing Distributions: Box plots are useful for comparing the distributions of multiple datasets or groups. They allow us to visually assess differences in central tendency, spread, and skewness.

3. Visualizing Data Distribution: Box plots provide central tendency, spread, and skewness of data. They show where the bulk of the data lies and whether it is skewed or symmetric.

4. Understanding Data Spread: The length of the box (interquartile range, IQR) represents the spread of the middle 50% of the data, providing insights into data variability.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
data=[box1,box2]
```

```python
plt.figure(figsize=(10, 8))

plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Values')
plt.ylabel('Box')
plt.title('Box Plot')

plt.tight_layout()
plt.show()
```
